<h1>Data loading and making chunks

```python
# Load csv file of dataset
trainData = pd.read_csv("/content/drive/MyDrive/Final Project - Fall 2021/Data/train.csv")
trainData = trainData.drop(["author", "recordist", "license", "rating", "date", "url", "file_type", "background", "primary_label", "description" ], axis=1)
trainData.head()
```

```python
# Calculating number of classes
uniqueLabels = trainData["ebird_code"].unique()
print(len(uniqueLabels))
```

```python
# Making folder for saving overlap chunk data
os.mkdir("/content/drive/My Drive/Final Project - Fall 2021/Overlap Chunk Data/")

# Making folders for each class under parent folder "Overlap Chunk Data"
for i in range(len(uniqueLabels)):
  os.mkdir("/content/drive/My Drive/Final Project - Fall 2021/Overlap Chunk Data/"+uniqueLabels[i])
```

```python
# Making overlap chunks of audios and saving them into classwise folders

errorFileId = []
sr = 44100

for id in range(0 ,len(trainData["filename"])):

  if (id%10 == 0 or id==(len(trainData["filename"])-1)):
    print("Processed Audios : ", id)
  
  if (id == len(trainData["filename"])):
    break;

  path = '/content/drive/MyDrive/Final Project - Fall 2021/Data/train_audio/'
  dir = trainData["ebird_code"][id]+'/'

  # Loading audio file
  x1, sr1 = librosa.load(path + dir + trainData["filename"][id], sr = sr)
  
  # Fetching audio duration from csv file
  audio_time = round(trainData["duration"][id])

  # Making overlap chunks of loaded audio
  for i in range(audio_time//3):
    t1 = i * 3 * 44100

    chunk = x1[t1:t1+(44100*5)]

    # Adding extra 0's if chunk duration is shorter than 5 seconds
    if(chunk.shape[0] != 220500):
      chunk = np.pad(chunk, (0, (44100*5 - chunk.shape[0])), 'constant')

    # saving audio chunks
    filename = '/content/drive/MyDrive/Final Project - Fall 2021/Overlap Chunk Data/'+trainData["ebird_code"][id]+'/'+trainData["filename"][id][0:-4]+'_'+str(i)+'.wav'
    sf.write(filename, chunk, 44100)
```
